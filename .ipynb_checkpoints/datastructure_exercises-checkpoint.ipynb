# 1. Distrowatch Hits Per Day
The website www.distrowatch.com provides details of the most popular Linux distributions, based on the number of hits per day (HPD). The details of the top 10 most popular distributions are as
follows:

| Distribution | H.P.D | 
| --- | --- |
| Debain | 1529 |
| Deepin | 844 |
| Elementary | 1312 |
| Fedora | 997 |
| MX Linux | 4914 |
| Manjaro | 2627 |
| Mint | 2049 |
| Solus | 1078 |
| Ubuntu | 1398 |
| Zorin | 827 |

Write a program which uses a loop to input the hits per day of the top 10 Linux distributions and
append them to a list. After the values have been added to the list, the program should sort the list
in reverse order and display it, and then calculate and display the total, average, median, maximum
and minimum hits per day.


Program design:

Input:

    - Number values (10 as per the linux distribution)
    - create an empty list of hpd

Processing:

    - initialise count variable to 0
    - while count < number_values:
          values = int(input("Enter the hits per days: "))
          hpd.append(values)
          
    - sort list in reverse order and display
    - calculate total, average, median, maximum, minimum

Output: 

    - display total, average, median, maximum and minimum

```python
# Version 1: while counting loop
# program response
print("Program to calculate the total, median, minimum and maximum for top 10 linux distribution.")
print("---"*30)

# input number values (10)
num_values = int(input("Enter the number value: "))

# create an empty list
hpd = []

# # intialise count variable to 0 
count = 0

# while counting loop
while count < 10:
    # input hit per days values
    hit_per_days = int(input("Enter the hits per days: "))
    # append hpd list 
    hpd.append(hit_per_days)
    # increment count by 1
    count += 1

print("---"*30)

# length of hpd
print(f"The total values in HPD is: {len(hpd)}")

# sort list into reverse order
hpd.sort(reverse=True)
print(f"The hpd (reverse order) is: {hpd}")

# calculating the total of HPD for linux distribution
print(f"The Total of HPD is: {sum(hpd)}")

# calculating the minimum and maximum for HPD
print(f"The maximum HPD is: {max(hpd)}, and the minimum HPD is: {min(hpd)}")

# calculating median of HPD -->> calculating the index for the moddle value than display the median
mid_index = int(len(hpd) / 2)
print("The index of middle value is: ", mid_index)
if len(hpd) % 2 == 1: # Odd number values
    median_hpd = hpd[mid_index]
else: 
    median = (hpd[mid_index-1] + hpd[mid_index]) / 2

# Display median value
print(f"The median value of HPD is: {median}")

# Calculating the average
print(f"The average of HPD is: {sum(hpd) / len(hpd):.1f}")
```

```python
# Version 2: Distrowatch hits per day -->> for loop
# program response
print("Program to calculate the total, median, minimum and maximum for top 10 linux distribution.")
print("---"*30)

# input number values (10)
num_values = int(input("Enter the number value: "))

# create an empty list
hpd = []

# input hits per day for each distribution
for i in range(num_values):
    hits_values = int(input("Enter the hits per day: "))
    hpd.append(hits_values)
    
print("---"*30)

# length of hpd
print(f"The total values in HPD is: {len(hpd)}")

# sort list into reverse order
hpd.sort(reverse=True)
print(f"The hpd (reverse order) is: {hpd}")

# calculating the total of HPD for linux distribution
print(f"The Total of HPD is: {sum(hpd)}")

# calculating the minimum and maximum for HPD
print(f"The maximum HPD is: {max(hpd)}, and the minimum HPD is: {min(hpd)}")

# calculating median of HPD -->> calculating the index for the moddle value than display the median
mid_index = int(len(hpd) / 2)
print("The index of middle value is: ", mid_index)
if len(hpd) % 2 == 1: # Odd number values
    median_hpd = hpd[mid_index]
else: 
    median = (hpd[mid_index-1] + hpd[mid_index]) / 2

# Display median value
print(f"The median value of HPD is: {median}")

# Calculating the average
print(f"The average of HPD is: {sum(hpd) / len(hpd):.1f}")
```

# 3: Standard Deviation of Pass Success Percentages
The standard deviation is a measure of the amount of variation of a set of values. A low standard
deviation indicates that the values tend to be close to the mean (average) of the values, while a high
standard deviation indicates that the values are spread out over a wider range.

Link: https://en.wikipedia.org/wiki/Standard_deviation

formula: s=√∑(x−¯x )2 / n−1

where x represents each individual value, x is the mean (average)
and n is the number of values. It calculates the variation as the
square root of the average squared deviation (x – x).

Write a program which inputs the pass success percentages, stores them in a list, and then calculates
and displays the average and the standard devation.

the total, divide by len -1 and then get the square root.

Program design:

Input:

    - create an empty list pass success rate
    
Processing:

    - while: Sentinel-controlled loop
        while True:
            input pass success percentage
            if choice is -1:
                break
            else:
                append the list
                
         if there are values in appended list then calculate
             - Mean
             - Standard deviation
             
Output:

    - display Mean and Standard Deviation

```python
print("*** version 1: Sentinel-controlled loop ***")

# importing the sqrt from math
from math import sqrt

# prgram response
print("Program to analyse the pass success percanetage (PS%).")
print("--"*30)

# create an empty list
ps_list = []

# using Sentinell-controlled loop
# input the PS% for each player
while True: 
    ps_rate = float(input("Enter the PS% for each player or -1 to finish: "))
    # if statement for finish choice
    if ps_rate == -1:
        break # break if -1 is the input
    else: 
        ps_list.append(ps_rate) # else store in the list

print() # empty space
print("---"*20) # decorator

# if statement to check if the list was appended
if len(ps_list) == 0:
    print("No values entered!")
else:
    # display the list
    print("The entered values of PS% for each player is: ", len(ps_list))
    
    # calculating & displaying the average
    mean = sum(ps_list) / len(ps_list)
    print(f"The average of PS% is: {mean:.1f}")
    
    print() # empty space
    
    # calculating and displaying the standard deviation
    print("*** Standard Deviation ***") # results
    print("--"*20)
    
    # list comprehension of deviation
    deviation_list = [(x-mean)**2 for x in ps_list] 
    
    # standard deviation
    std = sqrt(sum(deviation_list) / (len(ps_list) -1))
    
    # displaying the results
    print(f"The Standard deviation is: {std:.1f}")
```

```python
print("*** Version 2: for counting loop ***")

# importing the sqrt from math
from math import sqrt

# prgram response
print("Program to analyse the pass success percanetage (PS%).")
print("--"*30)

# input number values
num_values = int(input("Enter the number of values: "))

# create an empty list
ps_rate_list = []

# for counting loop
for i in range(num_values):
    ps_input = float(input("Enter the PS% for each player: "))
    ps_rate_list.append(ps_input)
    
print() # empty space

print("*** Mean & Standard Deviation ***")
print("---"*20)

# display the ps_rate_list
print("The total values of PS% are entered: ", len(ps_rate_list))

# calculating the mean
mean = sum(ps_rate_list) / len(ps_rate_list)

# displaying the mean
print(f"The mean of PS% is: {mean:.1f}")

# calculating the standard deviation
devation_list = [(x-mean)**2 for x in ps_rate_list] # list comprehension for deviation list
std_dev = sqrt(sum(deviation_list) / (len(ps_rate_list) - 1))

# displaying the std
print(f"The Standard Devistion is: {std_dev:.1f}")                      
```

# 4. Correlation of Player Passing Statistics
- Correlation is a measure of the relationship between 2 variables, x and y. Its value ranges from −1 to 1:
    - a correlation of 1 implies a perfect positive linear relationship between the variables
    - a correlation of -1 implies a perfect negative linear relationship between the variables
    - a correlation of 0 implies no relationship between the variables.
    
The formula is: r= ∑(x−¯x)( y−¯y) / √∑(x−¯x)2 * √∑( y−¯y )2

The following data, from WhoScored.com, provides the average number of passes per game and the pass success percentage (PS%) of players in the Premier League:

| | Name | Team, Age & Position | Pass per Game | PS% | 
| --- | --- | --- | --- | --- | 
| 1 | Aron Ramsdale | Bournemouth, 21, GK | 26.8 | 50.5 |
| 2 | Abdoulaye Doucouré | Watford, 26, M(C) | 62.4 | 84.0 |
| 3 | Adrián | Liverpool, 32, GK | 25.9 | 74.4 |
| 4 | Andre Gray | Watford, 28, FW | 9.1 | 71.2 |
| 5 | Andrew Robertson | Liverpool, 25, D(L),M(L) | 74.8 | 85.3 |
| 6 | Angus Gunn | Southampton, 23, GK | 23.8 | 41.7 |
| 7 | Ashley Barnes | Burnley, 29, AM(CL),FW | 15.8 | 61.1 |
| 8 | Ashley Westwood | Burnley, 29, DMC | 44.1 | 71.4 |

| 10 | Ben Godfrey | Norwich, 21, D(C) | 54.9 | 90.4 |

Source: https://www.whoscored.com/Regions/252/Tournaments/2/Seasons/7811/Stages/17590/PlayerStatistics/England-Premier-League-2019-2020

Do players who pass more have a higher pass success percentage?

Write a program which inputs the passes per game and the pass success percentage for each player,
stores the values in two separate lists, and then calculates and displays the correlation between them.

Hint:

For the numerator (top line): Use a list comprehension to generate the deviations for x, another list
comprehension to generate the deviations for y, and then multiply the corresponding values from the
two lists and store the result in a third list.

For the denominator (bottom line): Use a list comprehension to generate the squared deviations for
x, then use the sum function to calculate the total and then get the square root; do the same for y.

Program design:
    
Input: 

    - create an empty list 1). pass per game 2). pass success percentage
    
Processing:

    - While: Sentinel-Controlled loop
        while True:
            input the pass per game values
            if choice is -1:
                break
            else:
                append the list 2
                
        input the pass success percentage values
        append the list 1
    
    - if there are values in the appended list then
    - calculate Mean, deviation squared and correlation
                
Output:

    - display correlation

```python
print("*** Version: Sentinel-Controlled loop ***")

# Program response
print("Program to analyse the correlation between the pass per game and PS%.")
print("---"*25)

# importing the sqrt from math
from math import sqrt

# create the an empty lists 1). pass per game & 2). PS%
ppg = []
ps_rate = []

# while loop
while True:
    ppg_input = float(input("Enter the pas per game values for each player or -1 to finish: "))
    if ppg_input == -1:
        break
    else:
        ppg.append(ppg_input)
        
    ps_input = float(input("Enter the ps% for each player: "))
    ps_rate.append(ps_input)

    
print("---"*25)
print() # empty space

print("*** Correlation Analysis ***")
print("--"*15)

if len(ps_rate) == 0:
    print("No values entered!")
else:
    # calculate the mean
    ppg_mean = sum(ppg) / len(ppg)
    ps_rate_mean = sum(ps_rate) / len(ps_rate)
    
    # creating the deviation list for pass per game and pass success percentage
    x_deviation = [(x - ppg_mean) for x in ppg]
    y_deviation = [(y - ps_rate_mean) for y in ps_rate]
    
    # multiply both x and  y deviation list and create new single list
    xy_deviations = [x*y for x, y in zip(x_deviation, y_deviation)]
    
    # create the list of deviation sqaured
    x_sqr_deviation = [(x-ppg_mean)**2 for x in ppg]
    y_sqr_deviation = [(y-ps_rate_mean)**2 for y in ps_rate]
    
    # calculate the correlation
    correlation = sum(xy_deviations) / (sqrt(sum(x_sqr_deviation)) * (sqrt(sum(y_sqr_deviation))))
    
    # display the correlation
    print(f"The correlation is: {correlation:.2f}")
```

# 5:  Lotto Quick Pick
To win the Lotto, you must match the 6 winning numbers, which are between 1 and 47, inclusive.
Write 3 programs which simulate a "lotto quick pick". In each case, the program must randomly
selecting six unique numbers between 1 and 47, sort them in order and display them on screen. The
logic is provided here. As teams you must write the programs.

Program 1: Keep randomly selecting numbers until we get 6 different one

Program 2: Randomly select and remove 6 numbers from the list [1,2,3,...46,47]

Program 3: Shuffle the list [1,2,3,...46,47] and pick the first 6 numbers

Program Design:

Input: 

    - program 1:
    
        - create an empty list for quick pick
        - random numbers between 1-47 inclusively
        
    - program 2:
    
        - create an empty list for qucik pick
        - list containing randomly number generated 1-47 inclusively
        
    - program 3:
    
        - create an empty list fro quick pick 
        
Processing:

    - program 1:
    
        - while counting loop:
            while there are less than 6 numbers in the list
                randomly generate a number between 1 and 47
                    if number not in list
                        append the list
        - sort the list and display
        
    - program 2:
    
        - for repetition loop:
            for i in range(6)
                randomly choose the number from list
                remove the numbers
                append the list
        - sort and display the list
        
    - program 3:
    
        - shuffle the list containing numbers between 1 and 47
        - take the first 6 numbers from list and store them back into the list
        - sort and display the list
        
Output:

    - program 1, 2, & 3:
    
        - display the lotto quick pick numbers

```python
# program 1
print("*** Lotto qucik pick - program 1 ***")
print() # empty space

# inporting the randint from random
from random import randint

# program response
print("Program to display the Lotto quick pick numbers.")
print("--"*25)

print() # empty space

# create an empty list
number_list = []

# while repittion loop - while there are less than 6 numbers in the quick pick list
while len(number_list) < 6:
    # randomly generate the number between (1, 47)
    number = randint(1, 47)
    
    # if the number is not in the list
    if number not in number_list:
        # append the list
        number_list.append(number)
        
# sort the list and display
number_list.sort()
print("Your quick pick numbers are: ", number_list)
```

```python
# program 2
print("*** Lotto qucik pick - program 2 ***")
print() # empty space

# inporting the choice from random
from random import choice

# program response
print("Program to display the Lotto quick pick numbers.")
print("--"*25)

# create an empty list
number_list2 = []

# create a list containing the numbers 1-47 inclusive
all_numbers = list(range(1, 48))

# for repetition loop - repeat 6 times
for i in range(6):
    # randomly choose the number from the list - all_numbers
    number = choice(all_numbers)
    # remove the numbers from list - all_numbers
    all_numbers.remove(number)
    # add the number to the list
    number_list2.append(number)
    
# sort the list and display
number_list2.sort()
print("Your lotto quick pick numbers are: ", number_list2)
```

```python
# program 3
print("*** Lotto qucik pick - program 3 ***")
print() # empty space

# inporting the choice from random
from random import shuffle

# program response
print("Program to display the Lotto quick pick numbers.")
print("--"*25)

# create a list containing the numbers 1-47 number inclusively
all_numbers = list(range(1, 48))

# shuffle the list
shuffle(all_numbers)

# pick the first 6 number from the list
lotto_list = all_numbers[:6]

# sort a list and display 
lotto_list.sort()
print("Your lotto quick pick numbers are: ", lotto_list)
```

# 6: Letter Frequencies
Write programs to calculate letter frequencies of a piece of text. That is, the number of times each
letter of the alphabet (uppercase or lowercase) appears in the text. The logic is provided here. As
teams you must write the programs.

Only use the features that have been included in the course so far, e.g. do not use a dictionary.

Program 1: Using a for loop to process each letter in the text, one at a time

Program 2: Using a list comprehension to count the frequencies

Challenge: Extend your programs to display the letter with the highest frequency.

Program Design:

Input: 

    - ascii_lowercase import
    - text
    - create an empty list of frequency containing 26 elements, each initialising to 0
    
Processing:

    - input text
    - for loop to process each text - for each character in the text converted to lowercase
                - index of the letter in the lowercase alphabet
                - increase by 1 the value at the corresponding index of the frequencies list
    - list comprehension to process the each text - count the number of times each letter appears int eh text
    - calculate the most frequent letter
    - display the letter and frequency

Output:

    - Display letters and their frequency
    - Display the most frequent letter

```python
# program response
print("Program to calculate the text frequency from input texts.")
print("---"*25)
print() # empty space

print("*** Version 1: Using for loop to process each letter ***")
print("---"*30)

# importing ascii_lowercase from string
from string import ascii_lowercase

# input text
texts = input("Enter the texts: ")

# create a list - frequency containing 26 element - ascii_lowercase
frequencies = [0]*26

# for couting loop - for each character in text
for character in texts.lower():
    if character.islower():
        # find the index of the letter in lowercase alphabet
        index = ascii_lowercase.index(character)
        # increase by 1 the value at the corresponding index of the frequencies list
        frequencies[index] += 1
# display the text
# print(texts)

# header output
print("\nLetter  Frequency")

for i in range(26):
    print(f"  {ascii_lowercase[i]:<3}  {frequencies[i]:>5}")
    
# calculate the frequency
max_frequency = max(frequencies)
max_letter = ascii_lowercase[frequencies.index(max_frequency)]

# display the frequency
print("\nMost frequent letter is: ", max_letter)


print() # empty space
print("*** Version 2: Using list comprehension ***")
print("---"*25)

count_of_texts = [texts.lower().count(x) for x in ascii_lowercase] # using list comprehension to count the frequency

print("\nLetter  Frequency")
for i in range(26):
    print(f"  {ascii_lowercase[i]:<3}  {count_of_texts[i]:>5}")
    
# calculate the frequency
max_counts = max(count_of_texts)
max_letters = ascii_lowercase[count_of_texts.index(max_counts)]

# display the table
print("\nMost frequent letter is: ", max_letters)
```
