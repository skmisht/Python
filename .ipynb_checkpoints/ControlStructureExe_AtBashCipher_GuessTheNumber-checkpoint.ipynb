# 1: Guess the Number

Guess the Number is a simple guessing game. The computer “thinks” of a number, and the player
has ten attempts to guess it. If the player guesses the number, s/he wins. Otherwise, the computer
tells the player if the guess was too high or too low, and s/he can guess again.

Write a program to implement the game and test it. When you get the basic version working, design
and implement an improved version which repeatedly plays the game until the user wants to stop.

Hint: You'll need to import randint from the module random. To randomly generate a number between two numbers a and b inclusive, the code is randint(a,b).

```python
# impirting randint() from random 
from random import randint
# Program
print("Program to play the Guess the Number game.")
print("---"*20)
# program response
print("I'm thniking oa number between 1 and 100. Try to guess it..")
# Random number for computer pick (1, 100)
number = randint(1, 100)
# for loop to guess the number
for i in range(10):
    guess = int(input("Enter the guess: "))
    if guess == number: # if the guess is right
        print("You guessed it")
        choice = input("Play again (y/n)? ")
        if not choice.lower() == 'y':
            print("Game over!")
        break
    elif guess < number: # if the guess is too low
        print("Too low! Try again.")
    else: # if the guess is too high
        print("Too high! Try again.")
else:
    print("You are out of atempts (<10). Try again!")
```

# 2:  AtBash Cipher

A substitution cipher is a simple encryption method which replaces each letter in the original
message, called the plaintext, with another character in the enciphered (encrypted) message, called
the ciphertext. For example, the Atbash cipher substitutes each letter with the corresponding one at
the opposite end of the alphabet, i.e.

Write and test a Python program which implements the Atbash cipher, as follows:

    - Input the plaintext from the user
    - Create a variable to store the ciphertext, which should be initialised to an empty string ""
    - For each letter in the plaintext, find the corresponding letter at the opposite end of the alphabet and add this letter to the ciphertext
    - Display the ciphertext
    
When you get the basic version working, design, write and test an improved version which repeatedly inputs and enciphers the user's input, until s/he indicates that s/he is finished.

Hint: Import the variable ascii_lowercase from the module string. It contains the letters of the alphabet "abcdef...xyz"


```python
# importing the ascii_lowercase() from string method
from string import ascii_lowercase
# program response
print("Program using the Atbash Cipher ")
print("---"*20)

# while continuation until the user decide to quit.
while True:
    text = input("Enter the message to encipher or q to quit: ")
    ciphertext = ""
    if text =='q':
        break
    else:
        for letter in text.lower(): # for each letter in the text
            if letter.islower(): # if it's a lower
                number = ascii_lowercase.index(letter) # convert it to a number 
                new_number = (number + 1) % 26 # add to the number
                new_letter = ascii_lowercase[-new_number] # corresponding letter at the opposite end of the alphabet
                ciphertext += new_letter # add the letter to the ciphertext
            else: # just add it on
                ciphertext += letter
    # display the encrypted message
    print(f"The encrypted message is: {ciphertext}")
```
