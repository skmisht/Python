# 1: Basic Statistics
The file supercomputers.txt contains the performance of some
of the world's top supercomputers. The units are Teraflops, trillions
of calculations per second.

Write a program which will read the data from the file and then
calculate and display:

* the number of values
* the mean (average) value
* the maximum value
* the minimum value
* the median value
* and the standard deviation of the performances

Note: Use a list to store the data.

```python
# program response 
print("Program to calculate the basic statistics for supercomputing data.")
print("---"*20)
print()
# create an empty list
values_list = []
# input file
filename = input("Enter the filename: ")

# open the file
with open(filename) as input_file:
    for data in input_file:
#         data = float(data)
        values_list.append(float(data))
        

if len(values_list) == 0:
    print("No values to process.")
else:
    # total number of values
    print("The super_computing data: ", len(values_list))

    # mean/average value
    mean = sum(values_list) / len(values_list)
    print(f"The average value is: {mean:.2f} Tf")

    # Maximum and minimum value 
    print(f"The Maximum value is: {max(values_list)} Tf, and the Minimum value is: {min(values_list)} Tf")

    # Median value
    # first sort the data
    values_list.sort()
    # calculating the middle index
    mid_index = int(len(values_list) / 2)
    # display the middle index
    print("The middle index is: ", mid_index)
    # odd vs even
    if len(values_list) % 2 == 1:
        median = values_list[mid_index]
    else:
        median = (values_list[mid_index -1] + values_list[mid_index]) / 2
    
    # display the median value
    print(f"The Median value is: {median:.1f} Tf")

    # Calculating Standard Deviation and formula is -->> s=√∑(x−¯x )2 / n−1
    # deviation using the list comprehension 
    deviation = [(x-mean)**2 for x in values_list]
    # calulating the standard deviation
    from math import sqrt
    std = (sqrt(sum(deviation)) / (len(values_list) / 2))
    # display the standard deviation
    print(f"The Standard deviation is: {std:.1f} Tf")
```

# 3: FIFA: Correlation of Age and Player Ratings
The file fifa1.csv contains data on players in the following format:

| ID | Name | Age | Rating | 
| --- | --- | --- | --- | 

Write a program which:
* reads the header line (the first line of the file) using the .readline() method
* reads the remaining contents of the file line by line
* splits the line into the 4 components
* stores the ages in one list and the ratings in another list.
* displays the number of players and calculates and displays the correlation between age and rating.

```python
# Program response
print("Program to analyse FIFA data - Ages and Ratings.")
print("---"*25)
print()
# input the file
filename = input("Enter the filename: ")
# create an empty lists
age = []
ratings = []

# try to open the file
try:
    # open the file
    with open(filename, encoding="utf8") as input_file:
        # read first line - it contains the headers
        input_file.readline() # do not need store it
        
        # read in each line, one at a time
        for line in input_file:
            # try to split the line into ID, Name, Age and Ratings
            try:
                player_id, name, ages, rating = line.split(",")
            
                # add to the lists
                age.append(int(ages))
                ratings.append(int(rating))
                
            except ValueError:
                print("Line in incorrect format: ", line)
                
except FileNotFoundError:
    print("Unable to open file: ", filename)
    
if len(ratings) == 0:
    print("No values input from the file.")
else:
    # calculate the means
    age_mean = sum(age) / len(age)
    rating_mean = sum(ratings) / len(ratings)
    
    # calculating the correlation - The formula is: r= ∑(x−¯x)( y−¯y) / √∑(x−¯x)2 * √∑( y−¯y )2
    # create a list of the deviation - using list comprehension
    x_deviation = [(x-age_mean) for x in age]
    y_deviation = [(y-rating_mean) for y in ratings]
    
    # create a list of deviation multipled
    xy_deviations = [x*y for (x, y) in  zip(x_deviation, y_deviation)]
    
    # create a list of deviation squared 
    x_sqr_deviation = [ (x-age_mean)**2 for x in age]
    y_sqr_deviation = [ (y-rating_mean)**2 for y in ratings]
    
    # correlation
    correlation = sum(xy_deviations) / (sqrt(sum(x_sqr_deviation)) * (sqrt(sum(y_sqr_deviation))))
    
    # display the number of player & correlation
    print(f"Number of players: {len(ratings)}")
    print(f"Correlation: {correlation:.2f}")
```

# 4: Correlation of Value and Wages
The file fifa2.csv contains data on players in the following format:

| ID | Name | Value (£-M) | Wage (£-K) | 
| --- | --- | --- | --- | 

Write a program which reads the contents of the file line by line, splits the line into the 4
components and then stores the player values in one list and the wages in another list. You'll need to
strip the Euro sign and units M, K from the numbers before adding them to the lists, e.g.
* value = value.strip("€").strip("M").strip("K")
* wage = wage.strip().strip("€").strip("K")

The newline character has to be stripped from the contents of wage before stripping the signs.

Then display the number of players and calculate and display the correlation between player value
and wage.

```python
# program response
print("Program to analyse FIFA data - player'a values and wages.")
print("---"*20)
print()

# input the file
filename = input("Enter the filename: ")

# create an empty lists
wages = []
values = []

# try to open the file
try:
    with open(filename, encoding = 'utf8') as input_file:
        # read first line that contains the header
        input_file.readline() # do not need to store it
        
        # read the contents of file line by line 
        for line in input_file:
            
            try:
                player_id, name, value, wage = line.split(",")
                # striping the value and wages as the file contains the sylk extension
                # e.g. the data is- value: â‚¬110.5M, wage:â‚¬565K
                # strip("â").strip(",").strip("¬").
                # strip("â").strip(",").strip("¬")

                value = value.strip().strip("€").strip("M").strip("K")
                wage = wage.strip().strip("€").strip("K")
                # append into the lists
                values.append(float(value))
                wages.append(float(wage))
                
            except ValueError as value_error: # value error if the line isin incorrect format
                print(value_error)
                print("Line is in incorrect format: ", line)
            
except FileNotFoundError: # file not found error if there is no such a file
    print("Unable to find the file: ", filename)

if (len(values) == 0) or (len(wages) == 0):
    print("No values have been processed into the lists.")
else:
    print(f"Total number of players: {len(values)}")
    # calculating the means
    value_mean = sum(values) / len(values)
    wage_mean = sum(wages) / len(wages)
    
    # calculating the correlation and the formula is -->> r= ∑(x−¯x)( y−¯y) / √∑(x−¯x)2 * √∑( y−¯y )2
    # create alist of the deviations - using list comprehension
    from math import sqrt # sqrt module
    x_deviation = [(x-value_mean) for x in values]
    y_deviation = [(y-wage_mean) for y in wages]
    
    # create the list of deviation multiplied
    xy_deviations = [(x*y) for (x, y) in zip(x_deviation, y_deviation)]
    
    # create a list of the deviation squared
    x_sqr_deviation = [(x-value_mean)**2 for x in values]
    y_sar_deviation = [(y-wage_mean)**2 for y in wages]
    
    # Correlation
    correlation = sum(xy_deviations) / (sqrt(sum(x_sqr_deviation)) * sqrt((sum(y_sar_deviation))))
    
    # display correlation
    print(f"Correlation: {correlation:.2f}")
```

# 5: Text Analysis - Beowuff
The file beowulf.txt contains the contents of The Project Gutenberg EBook of Beowulf, can be found at: http://www.gutenberg.org/cache/epub/16328/pg16328.txt

Write a program which:
* opens the file beowulf.txt
* reads in its contents as a single string contents and displays
     - the number of characters
     - the number of words (using contents.split()) and
     - the number of lines (using contents.splitlines())
     
Then create a list (for example using a list comprehension) containing the lengths of the words, and
then calculate and display:
* the mean (average) word length
* the maximum word length
* the median word length
* the minimum word length
* the mode of the word lengths (the most common word length)

Note: To get the mode, create another list containing the frequencies of the word lengths, and find the
maximum value. 

```python
# program response
print("Program to analyse words length in Beowulf.")
print("---"*20)
print()

# input the file
filename = input("Enter the filename: ")

# try to open the file
try:
    # open the file
    with open(filename, encoding = 'utf8') as input_file:
        # read the contentsf of the file as a single string
        contents = input_file.read()
        
        # display the characters, words and lines
        print("Number of characters: ", len(contents))
        print("Number of words: ", len(contents.split()))
        print("NUmber of lines: ", len(contents.splitlines()))
        
        # creating the list of words - list comprehension
        words_length_list = [len(word) for word in contents.split()]
        
        # statistics on the words length
        print(f"Average word length: {sum(words_length_list) / len(words_length_list):.2f}")
        print(f"The maximum word length: {max(words_length_list)} and the Minimum word length: {min(words_length_list)}")
        
        # calculating the mode and for mode we need the frequency of the word length
        
        # the length of the longest word
        max_length = max(words_length_list)
        
        # list comprehension to create the frequencies of each word length
        frequencies = [(words_length_list.count(x)) for x in range(max_length+1)]
        
        # calculating the mode - word length with the highest frequency 
        max_frequency = max(frequencies)
        mode = frequencies.index(max_frequency)
        print(f"Most common word length: {mode}")
        
except FileNotFoundError as fnf_error:
    print(fnf_error)
    print("Unable to open the file: ", filename)
```

# 6. life.csv contains life expectancy data per country including
* GNP: Gross National Product, the value of all goods and services produced by the country.(* means no GNP value).
* Geographical Region: 1=Soviet Block/Warsaw Pact, 2=South America, 3=NATO/Other European, 4=Middle East, 5=Asia, 6=Africa
* Birth Rate: Number of births per 1,000 population
* Male Life Expectancy (LEXpM): The average numbers of years males are expected to live
* Female Life Expectancy (LEXpF): The average numbers of years females are expected to live

write a program to analyse the data and Include Exception Handling wherever appropriate.

```python
# program response
print("Program to analyse the life expectancy data for number of countries.")
print("---"*25)
print()

# input the file
filename = input("Enter the filename: ")

# create an empty lists 
countries = []
regions = []
gnps = []
birthrates = []
life_exp_M = []
life_exp_F = []

# try to open the file
try:
    with open(filename, encoding='utf8') as input_file:
        # read the first line that is header- column name
        col_names = input_file.readline()
        print(col_names)
        
        # for each line in the file
        for line in input_file:
            # try to process the line
            try:
                # split the line
                country, region, gnp, birth_rate, life_exp_Male, life_exp_Female = line.split(",")
            except ValueError as value_error: # invalid format to convert
                print(value_error)
                print("Line has invalid format so skipping: ", line)
                continue # continue to next line
                
            # append the lists - countries, region, gnp, birthrate, life_exp_M, life_exp_F
            countries.append(country)
            regions.append(region)
                
            # try to append the float data type
            try:
                gnps.append(float(gnp))
            except ValueError:
                print("Unable to read an invalid data type so skipping: ", gnp)
                    
            try:
                birthrates.append(float(birth_rate))
            except ValueError:
                print("Unable to read an invalid data type so skipping: ", birth_rate)
                
            try:
                life_exp_M.append(float(life_exp_Male))
            except ValueError:
                print("Unable to read an invalid data type so skipping: ", life_exp_Male)
                
            try:
                life_exp_F.append(float(life_exp_Female))
            except ValueError:
                print("Unable to read an invalid data type so skipping: ", life_exp_Female)
                
        print()
        # Analysis includes: 
        print("**** Number of Countries ****")
        print("---"*20)
        print(len(countries))
        
        # if the list has been append then go
        if len(countries) == 0:
            print("No values have been processed.")
        else:
            print("The number of countries: ", len(countries))
            
        print() # blank line
        if len(gnps) == 0:
            print("No values have been proccessed.")
        else:
            print("**** GNP analysis: Mean, Maximum, Minimum, Median, Range & Standard Deviation  ****")
            print("---"*25)
            # mean - gnp
            mean_gnp = sum(gnps) / len(gnps)
            print(f"The average gnp per country: {mean_gnp:.2f}")
            # maximum, minimum and range
            print(f"The maximum GNP: {max(gnps)} and the Minimum GNP: {min(gnps)}")
            print(f"The Range of GNP's (max-min): {max(gnps) - min(gnps)}")
            
            # median
            # middle index
            mid_index = int(len(gnps) / 2)
            
            if len(gnps) % 2 == 1: # odd number
                median = gnps[mid_index]
                print(f"Median GNP's: {median}")
            else:
                median = sum(gnps[mid_index-1 : mid_index + 1]) / 2
                print(f"Median GNP's: {median}")
                
            # std
            from math import sqrt
            # create the list of the deviation squared -list comprehension
            deviation = [(x-mean_gnp)**2 for x in gnps]
            # calculate the std
            std = sqrt(sum(deviation) / (len(gnps) - 1))
            print(f"The Standard deviation of GNPs: {std:.2f}")
            
            
        print() # blank line 
        if len(regions) == 0:
            print("No value have been processed.")
        else:
            # Mode - regions - most frequent region
            print("**** Most frequent Region ****")
            print("---"*20)
            
            # creating a list of the distinct values in order
            region_values = sorted(list(set(regions)))
            
            # count the number of times each region appears - list comprehension
            frequencies = [regions.count(x) for x in region_values]
            
            # find the index of most frequency region
            max_value = frequencies.index(max(frequencies))
            
            # display the max_value index i.e the region with the highest numbers of countires
            print(f"Region with most countires (Mode): {region_values[max_value]}")
            
            
        print() # blank line
        if len(birthrates) == 0:
            print("No values have been processed.")
        else:
            print("**** Correlation analysis *****")
            print("---"*20)
            
            # 1) correlation between Birth rate and life exp Male
            # calculating the mean first for both x = birthrate , y = life exp Male
            mean_birthrate = sum(birthrates) / len(birthrates)   
            mean_life_exp_Male = sum(life_exp_M) / len(life_exp_M)
            

            
            # create the list of deviations - list comprehension
            x_deviation = [(x-mean_birthrate) for x in birthrates]
            y_deviation = [(y-mean_life_exp_Male) for y in life_exp_M]
            
            # create the list of deviation multiplied - list comprehension
            xy_deviation = [(x*y) for (x, y) in zip(x_deviation, y_deviation)]
            
            # create a list of deviation squared - list comprehension
            x_sqr_deviation = [(x-mean_birthrate)**2 for x in birthrates]
            y_sqr_deviation = [(y-mean_life_exp_Male)**2 for y in life_exp_M]
            
            # correlation
            correlation = sum(xy_deviation) / (sqrt(sum(x_sqr_deviation)) * (sqrt(sum(y_sqr_deviation))))
            # display correlation
            print(f"Correlation between Life Expectancy Male (y) and Birth rate (x): {correlation:.2f}")
            
            # correlation between life exp Female and birthrate can be calculated same as Male
            
except FileNotFoundError as fnf_error:
    print(fnf_error)
    print("Unable to open the file: ", filename)
```
